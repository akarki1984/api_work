# Billboard: Hot 100 (Scraping to a CSV)

```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.billboard.com/charts/hot-100')
doc = BeautifulSoup(response.text)
```

```python
# doc
```

```python
rows = []

chart_list = doc.find_all(class_='chart-list-item')

for item in chart_list:
#     print(item ['data-rank'])
#     print(item['data-artist'])
#     print(item['data-title'])

    row = {}
    
    row['rank'] = item['data-rank']
    row['artist_name'] = item['data-artist']
    row['title'] = item['data-title']
    
    rows.append(row)
    
    
    rows
    
```

```python
rows
```

```python
df = pd.DataFrame(rows)
```
