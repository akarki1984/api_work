```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors')
doc = BeautifulSoup(response.text)
```

```python

ranks = doc.find_all('td', class_='number')

for rank in ranks:
    r=(rank.text.strip())
   
#     print(r)    
```

```python
book_name_list = doc.find_all('a', class_='bookTitle')

for book_name in book_name_list:
    b=(book_name.text.strip())
    
#     print(b)
```

```python
author_names = doc.find_all('a', class_='authorName')

for author_name in author_names:
    n = (author_name.text.strip())
    
   
```

```python
scores = doc.find_all('span', class_='smallText uitext')

for score in scores:
    s = (score.find('a').text)



# for score in scores:
#     print(score.text)
```

```python
votes = doc.find_all('span', class_='smallText uitext')

for vote in votes:
    v = (vote.find_all('a')[1].text)
    
#     print(v)
#     print(s)
    
```

```python
rows = []
ratings = doc.find_all('span', class_='minirating')

for rating in ratings:
    rt = (rating.text.strip())
    
     
    row = {}
    
    row['rank'] = r
    row['book_name'] = b
    row['author_name'] = n
    row['score'] = s
    row['vote'] = v
    row['rating'] = rt
    
    rows.append(row)

rows
    
    
   
```

```python
df = pd.DataFrame(rows)
```

```python
reads = []

for read in doc.find_all('tr'):

   book = {}

   book['rank'] = read.find('td', class_='number').get_text(strip=True)
   book['title'] = read.find('a', class_='bookTitle').get_text(strip=True)
   book['author'] = read.find('a', class_='authorName').get_text(strip=True)
   book['score'] = read.find('span', class_='smallText uitext').find("a").get_text(strip=True)
   #book['votes'] = read.find('span', class_='smallText uitext').find("a", [1]).get_text(strip=True)
#     book['rating'] = rating.text
   reads.append(book)

print(reads)
```

```python
df = pd.DataFrame(reads)
```
